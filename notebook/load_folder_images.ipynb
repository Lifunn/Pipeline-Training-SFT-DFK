```python
import shutil
from pathlib import Path

base     = Path("/content/drive/MyDrive/hf_images_folder/framevideo")
inner    = base / "framevideo"

if inner.exists():
    # Pindahkan semua isi inner ke base
    for item in inner.iterdir():
        dst = base / item.name
        shutil.move(str(item), str(dst))
        print(f"  ✓ Moved: {item.name}")

    # Hapus folder kosong
    inner.rmdir()
    print("\nSelesai ✓ Struktur sekarang:")
    for item in sorted(base.iterdir()):
        print(f"  {item.name}/")
else:
    print("Tidak ada double folder")
```

```python
import json
from pathlib import Path

IMAGE_DIR   = Path("/content/drive/MyDrive/hf_images_folder")
TRAIN_JSONL = "/content/drive/MyDrive/train_final_fixed.jsonl"
VAL_JSONL   = "/content/drive/MyDrive/valid_final_fixed.jsonl"

def check_coverage(jsonl_path, label):
    total        = 0
    text_only    = 0
    img_found    = 0
    img_missing  = 0
    missing_paths = []

    with open(jsonl_path) as f:
        for line in f:
            item = json.loads(line.strip())
            total += 1
            img_path = None

            for msg in item.get("messages", []):
                content = msg.get("content", "")
                if not isinstance(content, list):
                    continue
                for block in content:
                    if block.get("type") == "image" and block.get("image"):
                        img_path = block["image"]

            if img_path is None:
                text_only += 1
            else:
                if (IMAGE_DIR / img_path).exists():
                    img_found += 1
                else:
                    img_missing += 1
                    if len(missing_paths) < 5:
                        missing_paths.append(img_path)

    print(f"{'═'*55}")
    print(f"  {label}")
    print(f"{'═'*55}")
    print(f"  Total samples     : {total:,}")
    print(f"  Text-only         : {text_only:,} ({text_only/total*100:.1f}%)")
    print(f"  ✓ Image ditemukan : {img_found:,}")
    print(f"  ✗ Image tidak ada : {img_missing:,}")
    print(f"  Coverage          : {(text_only+img_found)/total*100:.1f}%")
    if missing_paths:
        print(f"\n  Contoh path tidak ditemukan:")
        for p in missing_paths:
            print(f"    {p}")

check_coverage(TRAIN_JSONL, "TRAIN")
print()
check_coverage(VAL_JSONL,   "VALIDASI")
```
